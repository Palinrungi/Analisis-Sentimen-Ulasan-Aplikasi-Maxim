```python
import os
import pandas as pd
from google_play_scraper import reviews_all, Sort

# Tentukan nama file untuk menyimpan/memuat data
csv_file = 'ulasan_aplikasiMaxime.csv'

# Cek apakah file CSV sudah ada
if os.path.exists(csv_file):
    # Jika file ada, langsung baca dari CSV
    print(f"File '{csv_file}' ditemukan. Memuat data dari file...")
    app_reviews_df = pd.read_csv(csv_file)
else:
    # Jika file tidak ada, lakukan scraping
    print(f"File '{csv_file}' tidak ditemukan. Memulai proses scraping...")
    scrapreview = reviews_all(
        'com.taxsee.taxsee',
        lang='id',
        country='id',
        sort=Sort.MOST_RELEVANT,
        count=1000
    )
    
    # Konversi hasil scraping ke DataFrame
    app_reviews_df = pd.DataFrame(scrapreview)
    
    # Simpan DataFrame ke file CSV untuk penggunaan selanjutnya
    app_reviews_df.to_csv(csv_file, index=False)
    print(f"Data hasil scraping telah disimpan ke '{csv_file}'.")

# Tampilkan informasi dan beberapa baris pertama dari DataFrame
print("\nInformasi DataFrame:")
app_reviews_df.info()

print("\n5 baris pertama data:")
app_reviews_df.head()
```
